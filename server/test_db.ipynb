```python
import sqlite3

DB_PATH = "../data/olist.sqlite"

conn = sqlite3.connect(DB_PATH)

cursor = conn.cursor()
```

```python
cursor.execute(
    "SELECT name FROM sqlite_master WHERE type='table';"
)

tables = cursor.fetchall()

tables
```

```python
cursor.execute("PRAGMA table_info(orders);")

cols = cursor.fetchall()

cols

for c in cols:
    print(c[1], "-", c[2])
```
